# Question 1 : Create an object named student with the following properties:
● name: "Rahul"

● age: 21

● courses: an array containing "Math", "Physics", and "Chemistry"

Then, perform the following operations:
1. Add a new property grade with the value "A".
2. Update the age to 22.
3. Delete the courses property.

Ans 1:

Here is how to create and manipulate the student object in JavaScript according to the question:
<pre>
// Step 1: Create the object named student
const student = {
  name: "Rahul",
  age: 21,
  courses: ["Math", "Physics", "Chemistry"]
};

// Step 2: Add a new property grade with value "A"
student.grade = "A";

// Step 3: Update the age to 22
student.age = 22;

// Step 4: Delete the courses property
delete student.courses;

// student object after all operations
console.log(student);
</pre>

Explanation:

- At first, the student object is created with three properties: name, age, and courses.

- Then, using dot notation, a new property grade is added.

- The age property is updated by direct assignment.

- The courses property is removed using the delete operator.

After these operations, the student object will look like this:
<pre>
{
  name: "Rahul",
  age: 22,
  grade: "A"
}
</pre>

# Question 2: Write a function checkEligibility that takes an age as a parameter and returns:
● "Eligible" if age is 18 or above

● "Not Eligible" if age is below 18

Use an if-else statement.


Ans 2:

Here is the JavaScript function checkEligibility that checks the age and returns the eligibility status using an if-else statement:
<pre>
function checkEligibility(age) {
  if (age >= 18) {
    return "Eligible";
  } else {
    return "Not Eligible";
  }
}
</pre>
Explanation:

- The function takes one parameter age.

- It uses an if-else statement to compare the age with 18.

- If the age is 18 or more, the function returns "Eligible".

- Otherwise, it returns "Not Eligible".

You can test the function like this:
<pre>
console.log(checkEligibility(20)); // Output: Eligible
console.log(checkEligibility(16)); // Output: Not Eligible
</pre>

# Question 3: Write a function getDayName that takes a number (1-7) and returns the corresponding day of the week. Use a switch statement.


Ans 3:

Here is the JavaScript function getDayName that takes a number from 1 to 7 and returns the corresponding day of the week using a switch statement:
<pre>
function getDayName(dayNumber) {
  switch(dayNumber) {
    case 1:
      return "Monday";
    case 2:
      return "Tuesday";
    case 3:
      return "Wednesday";
    case 4:
      return "Thursday";
    case 5:
      return "Friday";
    case 6:
      return "Saturday";
    case 7:
      return "Sunday";
    default:
      return "Invalid day number";
  }
}
</pre>
Explanation:

- The function takes one parameter dayNumber.

- The switch statement checks the value of dayNumber.

- For each case (1 to 7), it returns the corresponding day name.

- If the number is not between 1 and 7, the default case returns "Invalid day number".

You can test the function like this:
<pre>
console.log(getDayName(3)); // Output: Wednesday
console.log(getDayName(7)); // Output: Sunday
console.log(getDayName(10)); // Output: Invalid day number
</pre>

# Question 5: Given an array numbers = [1, 2, 3, 4, 5], perform the following:
1. Use a for loop to print each number.
2. Use a while loop to print each number.
3. Use a for-in loop to print each index.
4. Use a for-of loop to print each number.

Ans 5:

Here is how to perform the tasks on the array numbers = [1, 2, 3, 4, 5] using different loops:
<pre>
const numbers = [1, 2, 3, 4, 5];

// 1. Use a for loop to print each number
for (let i = 0; i < numbers.length; i++) {
  console.log(numbers[i]);
}

// 2. Use a while loop to print each number
let j = 0;
while (j < numbers.length) {
  console.log(numbers[j]);
  j++;
}

// 3. Use a for-in loop to print each index
for (let index in numbers) {
  console.log(index);
}

// 4. Use a for-of loop to print each number
for (let num of numbers) {
  console.log(num);
}
</pre>
Explanation:

- The for loop uses an index to iterate and print values.

- The while loop uses a counter variable to iterate until condition fails.

- The for-in loop iterates over the indices (keys) of the array.

- The for-of loop iterates over the values directly in the array.

All four loops output the required elements or indices as specified.

# Question 6: Explain what functions are in JavaScript and provide an example of a function that adds two numbers.

Ans 6:

Functions in JavaScript are reusable blocks of code designed to perform a specific task or calculate a value when called. They help organize code into modular pieces so you can write code once and execute it multiple times with different inputs.

Functions are defined using the function keyword, followed by a name, parentheses (which may include parameters), and curly braces containing the code to execute. When you call a function, it runs the code inside and can return a result.

Here’s a simple example of a function that adds two numbers:
<pre>
function addNumbers(a, b) {
  return a + b;
}

// Example usage:
console.log(addNumbers(3, 5)); // Output: 8
</pre>
Explanation:

- function addNumbers(a, b) declares a function named addNumbers with two parameters a and b.

- The function returns the sum of a and b.

- Calling addNumbers(3, 5) outputs 8 to the console.

Functions are fundamental in JavaScript for structuring logic, handling events, and managing code repetition efficiently.

# Question 7: Convert the add function from Question 6 into an arrow function.

Ans 7:

To convert the addNumbers function from Question 6 into an arrow function, here is the equivalent JavaScript code:

Regular function:
<pre>
function addNumbers(a, b) {
  return a + b;
}
</pre>
Arrow function version:
<pre>
const addNumbers = (a, b) => a + b;
</pre>
Explanation:

- The function keyword is replaced by a constant variable assignment using const.

- The parameters go inside parentheses (a, b).

- The arrow => follows the parameters.

- Since the function only returns an expression, the curly braces and return keyword are omitted for brevity.

- This makes the function more concise without changing its behavior.

You can use it like this:
<pre>
console.log(addNumbers(4, 6)); // Output: 10
</pre>
This conversion is a common practice in modern JavaScript for cleaner and shorter function expressions.

# Question 8: Write a function greet that takes a name and a callback function. The greet function should call the callback with the name.

Ans 8:

Here is the JavaScript function greet that takes a name and a callback function, then calls the callback with the name:
<pre>
function greet(name, callback) {
  callback(name);
}
</pre>
Example usage:
<pre>
function sayHello(name) {
  console.log("Hello, " + name + "!");
}

greet("Rahul", sayHello); // Output: Hello, Rahul!
</pre>
Explanation:

- The greet function accepts two parameters: a string name and a callback function.

- Inside greet, it calls the callback function, passing the name as an argument.

- The callback function sayHello receives the name and prints a greeting message.

This demonstrates how to use callback functions to achieve customizable behavior in JavaScript.

# Question 9: Given the array: const fruits = ["apple", "banana", "cherry"];
1. What is the length of the array?
2. Add "date" to the end of the array using the length property.


Ans 9:
1. The length of the array fruits is 3, since it contains three elements: "apple", "banana", and "cherry".

2. To add "date" to the end of the array using the length property, you can assign "date" to the index equal to the current length of the array:
<pre>
const fruits = ["apple", "banana", "cherry"];

// Length of the array
console.log(fruits.length); // Output: 3

// Add "date" to the end using the length property
fruits[fruits.length] = "date";

console.log(fruits); // Output: ["apple", "banana", "cherry", "date"]
</pre>
Explanation:

- The .length property returns the number of elements in the array, which is 3 here.

- Using fruits.length as the index appends the new element "date" at the next available position.

- This effectively adds "date" as the fourth element of the array.

This approach leverages the dynamic size of arrays in JavaScript to add elements efficiently.

# Question 10: Given the string text = " Hello World! ", perform the following:
1. Trim the whitespace.
2. Convert to uppercase.
3. Check if it includes "World".


Ans 10:

Here is how to perform the required operations on the string text = " Hello World! " in JavaScript:
<pre>
const text = " Hello World! ";

// 1. Trim the whitespace
const trimmedText = text.trim();
console.log(trimmedText); // Output: "Hello World!"

// 2. Convert to uppercase
const upperText = trimmedText.toUpperCase();
console.log(upperText); // Output: "HELLO WORLD!"

// 3. Check if it includes "World"
const containsWorld = trimmedText.includes("World");
console.log(containsWorld); // Output: true
</pre>
Explanation:

- .trim() removes whitespace from both ends of the string.

- .toUpperCase() converts the string to uppercase letters.

- .includes("World") checks if the substring "World" is present, returning true or false.

This completes the string manipulations as requested for Question 10.

# Question 11: Given the array numbers = [1, 2, 3, 4, 5], perform the following:
1. Add 6 to the end.
2. Remove the first element.
3. Find the index of 4.
4. Check if all numbers are greater than 0.


Ans 11:

Here are the operations performed on the array numbers = [1, 2, 3, 4, 5] in JavaScript:
<pre>
const numbers = [1, 2, 3, 4, 5];

// 1. Add 6 to the end
numbers.push(6);
console.log(numbers); // Output: [1, 2, 3, 4, 5, 6]

// 2. Remove the first element
numbers.shift();
console.log(numbers); // Output: [2, 3, 4, 5, 6]

// 3. Find the index of 4
const indexOf4 = numbers.indexOf(4);
console.log(indexOf4); // Output: 2

// 4. Check if all numbers are greater than 0
const allGreaterThanZero = numbers.every(num => num > 0);
console.log(allGreaterThanZero); // Output: true
</pre>
Explanation:

- push(6) adds 6 to the end of the array.

- shift() removes the first element (originally 1).

- indexOf(4) returns the index where 4 is found, which is 2 after removal of first element.

- every(num => num > 0) checks if all elements satisfy the condition of being greater than 0 and returns a boolean.

This completes all the requested array manipulations for Question 11.
